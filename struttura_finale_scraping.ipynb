```python
import pandas as pd
```

```python
tweets_netflix = pd.read_csv("/content/drive/Shareddrives/information project/ANALISI X (Twitter)/Fase 2 - Scraping twitter/tweets_netflix.csv")
tweets_svb = pd.read_csv("/content/drive/Shareddrives/information project/ANALISI X (Twitter)/Fase 2 - Scraping twitter/tweets_svb.csv")

print("=== Netflix ===")
print(tweets_netflix.head())
print("\nColonne:", tweets_netflix.columns.tolist())

print("\n=== SVB ===")
print(tweets_svb.head())
print("\nColonne:", tweets_svb.columns.tolist())
```

```python
# Convertiamo la colonna 'date' in datetime
for df in [tweets_netflix, tweets_svb]:
    df['date'] = pd.to_datetime(df['date'])

# Funzione per assegnare il giorno di trading USA
def assign_us_trading_day(date):
    weekday = date.weekday()  # Lun=0 ... Dom=6
    if weekday == 5:  # Sabato
        return date + pd.Timedelta(days=2)  # Luned√¨
    elif weekday == 6:  # Domenica
        return date + pd.Timedelta(days=1)  # Luned√¨
    else:
        return date  # Lun-Ven stesso giorno

# Applichiamo la funzione
for df in [tweets_netflix, tweets_svb]:
    df['us_trading_day'] = df['date'].apply(assign_us_trading_day)

# Controllo rapido
print("=== Netflix ===")
print(tweets_netflix[['date','us_trading_day']].head(10))

print("\n=== SVB ===")
print(tweets_svb[['date','us_trading_day']].head(100))
```
