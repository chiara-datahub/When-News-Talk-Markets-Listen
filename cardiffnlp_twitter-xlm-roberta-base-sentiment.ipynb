```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
```

```python

input_path = "/content/drive/Shareddrives/information project/ANALISI X (Twitter)/Fase 2 - Scraping twitter/tweets_netflix.csv"

tweets_df = pd.read_csv(input_path)

# Modello multilingua per tweet
model_name = "cardiffnlp/twitter-xlm-roberta-base-sentiment"

# Tokenizer e modello
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Creiamo pipeline sentiment
sentiment_pipeline = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

# Funzione per calcolare sentiment score numerico
def get_sentiment_score(text):
    result = sentiment_pipeline(text[:512])[0]  # taglia a 512 token se troppo lungo
    label = result['label']  # POSITIVE / NEUTRAL / NEGATIVE
    score = result['score']
    # Convertiamo in numerico: positive=1, neutral=0, negative=-1 * score
    if label.lower() == "positive":
        return score
    elif label.lower() == "negative":
        return -score
    else:
        return 0.0

# Applichiamo al dataset
tweets_df['xlm_score'] = tweets_df['text'].apply(get_sentiment_score)

```

```python
# Salviamo CSV con punteggi multilingua
output_path = "/content/drive/Shareddrives/information project/ANALISI X (Twitter)/Fase 3 - Sentiment analysis/tweet_netflix_xlm.csv"
tweets_df.to_csv(output_path, index=False)

print("CSV con sentiment multilingua salvato in:")
print(output_path)
```
