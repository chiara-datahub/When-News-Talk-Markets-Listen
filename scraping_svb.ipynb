```python
import pandas as pd
from datetime import timedelta
import requests
import time
```

```python
API_TOKEN = "token"

URL = (
    "https://api.apify.com/v2/acts/"
    "kaitoeasyapi~twitter-x-data-tweet-scraper-pay-per-result-cheapest/"
    "run-sync-get-dataset-items"
    f"?token={API_TOKEN}"
)
```

```python
start_date = pd.to_datetime("2023-03-10")
end_date   = pd.to_datetime("2023-03-17")

dates = pd.date_range(start=start_date, end=end_date, freq="D")
```

```python
payloads_svb = []

for d in dates:
    since = d.strftime("%Y-%m-%d_00:00:00_UTC")
    until = (d + timedelta(days=1)).strftime("%Y-%m-%d_00:00:00_UTC")

    payload = {
        "twitterContent": "Silicon Valley Bank banking system",
        "since": since,
        "until": until,
        "lang": "en",
        "queryType": "Top",
        "maxItems": 100
    }

    payloads_svb.append(payload)
```

```python
all_tweets = []

for payload in payloads_svb:
    print("Scarico:", payload["since"], "‚Üí", payload["until"])

    response = requests.post(URL, json=payload)
    tweets = response.json()

    if isinstance(tweets, list) and len(tweets) > 0:
        all_tweets.extend(tweets)

    time.sleep(2)
```

```python
tweets_df = pd.DataFrame(all_tweets)
```

```python
tweets_clean = tweets_df[tweets_df["type"] == "tweet"].copy()
cols_to_keep = [
    "id",
    "text",
    "twitterUrl",
    "createdAt",
    "lang",
    "retweetCount",
    "replyCount",
    "likeCount",
    "quoteCount",
    "isReply",
    "isQuote"
]

tweets_clean = tweets_clean[cols_to_keep]
```

```python
import pandas as pd

tweets_clean["createdAt"] = pd.to_datetime(
    tweets_clean["createdAt"], errors="coerce"
)

tweets_clean["date"] = tweets_clean["createdAt"].dt.date
```

```python
tweets_clean = tweets_clean.drop_duplicates(subset="id")
tweets_clean.reset_index(drop=True, inplace=True)
```

```python
cols_to_drop = [col for col in ['isQuote', 'lang', 'createdAt'] if col in tweets_clean.columns]
tweets_clean = tweets_clean.drop(columns=cols_to_drop)
```

```python
tweets_clean = tweets_clean[tweets_clean['text'].str.match(r'^[\x00-\x7F]+$', na=False)]

```
